### 실전 : 여러 조건에 맞춰 수집하기

```python
query = 'original' # query, bookName, title, original, translation, sprint, etc 중 선택
keyword = '仁' # 검색어 
page = 1 # 수집 시작 페이지
ipp = 1000 # 한페이지에 표시할 검색 결과수
url = f'http://db.cyberseodang.or.kr/front/openapi/sabu.do?{query}={keyword}&page={page}&ipp={ipp}'
```

```python
req_str = requests.get(url).text
```

```python
# 검색자료 총 검색건수(total)
ET.fromstring(req_str).find('.//info//total').text
```

```python
# 검색자료 총 페이지수(pageCnt)
pageCnt = int(ET.fromstring(req_str).find('.//info//pageCnt').text)
pageCnt
```

```python
row_list = []
for num in tqdm(range(pageCnt)):
  url = f'http://db.cyberseodang.or.kr/front/openapi/sabu.do?{query}={keyword}&page={num+1}&ipp={ipp}'
  
  req_str = requests.get(url).text
  xtree = ET.fromstring(req_str).findall('.//items//item')

  for nodes in xtree:
    field = {}
    field['id'] = nodes.findtext('id')
    field['title'] = nodes.findtext('title')
    field['content'] = nodes.findtext('content')
    field['url'] = nodes.findtext('url')
    row_list.append(field)
```

```python
df = pd.DataFrame(row_list)
df
```
