####elif 활용한 if문
"주머니에 돈이 있으면 택시를 타고, 주머니에 돈은 없지만 카드가 있으면 택시를 타고, 돈도 없고 카드도 없으면 걸어 가라."

##### 중복 삭제(drop_duplicates)

```
data = {'key1':['a', 'b', 'b', 'c', 'c'],
'key2':['v', 'w', 'w', 'x', 'y'],
'col':[1, 2, 2, 4, 5]}
df = pd.DataFrame(data,columns=['key1','key2','col'])
df
```

```
df.duplicated(['key2']) # key2의 column에서 중복이 있는지 없는지 불 형태로 표현
```

```
df.duplicated(['key1', 'key2']) #key1, key2 둘 다 중복이 된 것 만 표현
```

```
df.drop_duplicates(['key1'], keep='first') #key1을 기준으로 중복되는 부분의 행을 지우는데 처음 것을 살려라
```

```
df.drop_duplicates(['key1'], keep='last')
```

```
df.drop_duplicates() #아예 똑같은 행을 지움
```

```
df.dropna()
```
